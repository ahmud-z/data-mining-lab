```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_visualization.csv")
```

```python
sns.barplot(x='Gender', y='Salary', data=df)
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True, bins=10)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Salary', 'Experience']])
plt.suptitle('Pairplot of Age, Salary, and Experience', y=1.02)
plt.show()
```
