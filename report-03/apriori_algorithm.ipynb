```python
import pandas as pd
from mlxtend.preprocessing import  TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df =pd.read_csv('transactions.csv')
df.head()
```

```python
print("Column Name  Total Null")
df.isnull().sum()
```

```python
df.drop("Transaction", axis=1, inplace=True)
df.head()
```

```python
df.fillna("", inplace=True)
df.values.tolist()
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(df.values.tolist())
te_array[:3]
```

```python
te.columns_
```

```python
transaction_df = pd.DataFrame(te_array, columns=te.columns_)
transaction_df.head()
```

```python
freq_items = apriori(transaction_df, min_support=0.6, use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items, min_threshold=0.8)
rules
```

```python
rules[['antecedents', 'consequents', 'support', 'confidence']]
```
